# **Least Square Fit** 

```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy.linalg import solve

x=np.array([1,2,3,4,5])
y=np.array([3.1,5.9,7.2,10.0,11.3]) 

# Computing the sums

n   = x.size 
sx  = np.sum(x)
sy  = np.sum(y)
sx2 = np.sum(x**2)
sx3 = np.sum(x**3) 
sx4 = np.sum(x**4) 
sxy = np.sum(x*y)
sx2y = np.sum((x**2)*y) 

# construct matrices 
# Linear Fit 
A = np.array([[n,sx],[sx,sx2]]) 
B = np.array([sy,sxy]) 

# Quadratic Fit 
AA = np.array([[n,sx,sx2],[sx,sx2,sx3],[sx2,sx3,sx4]]) 
BB = np.array([sy,sxy,sx2y]) 
              
# Solutions of equations using linear algebra module 
# Linear 
C = solve(A,B) 
a0,a1 = C

# Quadratic 
CC = solve(AA,BB) 
aa0,aa1,aa2 = CC

print(a0,a1) 
print(aa0,aa1,aa2) 
# Plot 
yfit = a0 + a1*x
yyfit = aa0 + aa1*x + aa2*x**2

plt.plot (x,y, 'og') 
plt.plot(x,yfit,'blue',label='Linear') 
plt.plot(x,yyfit,'red',label='Quadratic') 

plt.xlabel('x') 
plt.ylabel('y') 
plt.legend() 

plt.show() 
```
