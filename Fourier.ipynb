## Sawtooth wave

```python
import numpy as np
from scipy.signal import sawtooth
import matplotlib.pyplot as plt 

from scipy.integrate import quad 

x = np.linspace(0, 2*np.pi, 100)

def sawt(x):
    return sawtooth(x,width=0.5)

y = sawt(x) 

fc=lambda x:sawt(x)*cos(i*x) 
fs=lambda x:sawt(x)*sin(i*x)

An=[] 
Bn=[] 

n=8
i=0
a0=quad(fc,-np.pi,np.pi)[0]*(1/np.pi)
sum=a0/2

for i in range(n):
    an=quad(fc,-np.pi,np.pi)[0]*(1/np.pi) 
    An.append(an) 
    bn=quad(fs,-np.pi,np.pi)[0]*(1/np.pi) 
    Bn.append(bn) 

for i in range(n):
    sum= sum + An[i]*np.cos(i*x) + Bn[i]*np.sin(i*x) 

plt.plot(x,sum,'r--') 

plt.plot(x,y,'b')
plt.show() 
```
