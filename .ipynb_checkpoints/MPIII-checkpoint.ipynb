## **Solution of ODE using Shooting Method** 

### **Example 1** : The ODE we are trying to solve
$$\frac{d^2y}{dx^2} = - g$$ 
 

> with  $洧녮=9.8$ and Boundary Conditions to be  $洧녽(洧논=0)=0, ~洧녽(洧논=10)=50$ 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt  

def f(u,x):
    y, z = u
    f1, f2 = z, -9.8
    return [f1,f2]

x=np.linspace(0,10,50) 
y0 = 0
y1 = 50
z1, z2 = 50, 60

u = [y0,z1]
sol = odeint(f,u,x) 
w1 = sol[:,0][-1] 

u1=[y0,z2]
sol1=odeint(f,u1,x)

plt.plot(x,sol[:,0],'b') 
plt.plot(x,sol1[:,0],'g') 

tol = 0.01

for i in range(1000):
    u = [y0, z2] 
    sol = odeint(f,u,x)
    w2 = sol[:,0][-1]
    print("Iteration : ",i, w1, w2 )     
    if abs(w1 - w2) < tol:
        break 
    z1, z2 = z2, z2 + (z2 - z1)/(w2 - w1)*(y1 - w2) 
    w1 = w2 

    
#print(no_of_iterations) 

# plotting 
plt.grid() 

plt.plot(x,sol[:,0],'o')
plt.plot(10,y1,marker='o',markersize=20)  
plt.show() 
```

### **Example 2** : The ODE we are trying to solve (SHM)
$$\frac{d^2y}{dx^2} = - y$$ 
 

> with Boundary Conditions to be  $洧녽(洧논=0)=0, ~洧녽(洧논=\pi) = 1$ 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt  

def f(u,x):
    y, z = u
    f1, f2 = z, -y 
    return [f1,f2]

x=np.linspace(0,10,50) 
y0 = 1
y1 = 1
z1, z2 = 50, 60

u = [y0,z1]
sol = odeint(f,u,x) 
w1 = sol[:,0][-1] 

u1=[y0,z2]
sol1=odeint(f,u1,x)

plt.plot(x,sol[:,0],'b') 
plt.plot(x,sol1[:,0],'g') 

tol = 0.0000001

for i in range(1000):
    u = [y0, z2] 
    sol = odeint(f,u,x)
    w2 = sol[:,0][-1]
    print("Iteration : ",i, w1, w2 )         
    if abs(w1 - w2) < tol:
        break 
    z1, z2 = z2, z2 + (z2 - z1)/(w2 - w1)*(y1 - w2) 
    w1 = w2 
    print(i) 

# plotting 
plt.grid() 

plt.plot(x,sol[:,0],'o')
plt.plot(10,y1,marker='o',markersize=20)  
plt.show() 
```

## **Solution of ODE using Finite Difference Method** 

### **Example 1** : The ODE we are trying to solve 

$\frac{d^2y}{dx^2} = -g $ 

with $g = 9.8$ and Boundary Conditions to be $y(x=0) = 1, y(x=5) = 50$

```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt 

x0=0
xf=100
y0=0
yf=20
n=100 
y[0]=y0
y[100]=yf
x,h = np.linspace(x0,xf,n+1,retstep=True) 

print(x,h) 


y = np.zeros(n+1)


y[0],y[100]=0,30
err = 100

ctr=0
N=5000
fig = plt.figure() 
ax=fig.add_subplot(111) 
while err > 1e-4 :
##while i < 500: 
    y0=y.copy() 
    for i in range(1,n):
        y[i] = (y[i+1] + y[i-1])/2 + 9.8*h**2/2 
    err = norm(y-y0)
    if ctr%1000 == 1:
        line1=ax.plot(x,y)
        fig.canvas.draw() 
    ctr = ctr + 1 
    
# Plotting 

plt.show()
```

### **Example 2** : The ODE we are trying to solve 

$y''(x) + y(x) = 0$ 

Boundary Conditions to be $y(x=0) = 1$ and $y(x = \pi)=1$ for $0<x<\pi$ 

```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt 
x,h = np.linspace(0,1,101,retstep=True) 
y = np.zeros(101)

y[0],y[100]=0,1
err = 100

n=0
N=5000
fig = plt.figure() 
ax=fig.add_subplot(111) 
while err > 1e-4 :
#while n < 500: 
    y0=y.copy() 
    for i in range(1,100):
        y[i] = (y[i+1] + y[i-1])/(2 - h**2) 
    err = norm(y-y0)
    n=n+1
    if n%1000 == 1:
        line1=ax.plot(x,y)
        fig.canvas.draw() 
    n = n + 1 
# Plotting 

plt.show()
```

```python
# Finite Difference Method : Solution using Matrix Method 

import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse 
plt.style.use('seaborn-poster')

pi=22/7

steps = 10
x0=0
xf=pi
y0=1
yf=1
n = steps
h = (xf-x0) / (n+1) 
print('h =',h) 
# Get A
A = np.zeros((n-1,n-1))

a1=-(2 - h**2)*np.ones(n)
a2=np.ones(n-1) 
A=sparse.diags([a1,a2,a2],(0,-1,1)).todense() 

#diagonals = [[1,2,3,4], [1,2,3], [1,2]]
#d1=sparse.diags(diagonals, [0, -1, 2])
#print(d1) 

# Get b
b = np.zeros(n)
b[0] = -1
b[-1] = -1

# solve the linear equations
y = np.linalg.solve(A, b)
t = np.linspace(0, pi, n)

plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.show()
```

# **Integration by Quadratures** 

## **Gauss-Legendre Quadratures** 

**Syntax** 

scipy.integrate.quadrature(func, a, b, args = (), 
                           tol = 1.49e-08, 
                           rtol = 1.49e-08, 
                           maxiter = 50, 
                           vec_func = True, 
                           miniter = 1)

Parameters
func: This is the Python function or the method that is being integrated.

a: This is a float value that specifies the lower limit of integration.

args: This is an optional tuple parameter. It specifies the extra arguments to be passed to the func function.

tol, rtol: This is an optional float value that defines the absolute tolerance.

Note: Iteration stops when the error between the last two iterates is less than tol or the relative change is less than rtol. Its default value is 1.49e-08.

maxiter: This is an optional int value that specifies the maximum order of the Gaussian quadrature.

miniter: This is an optional int value that specifies the minimum order of the Gaussian quadrature.

vec_func: This is an optional bool value that specifies True or False, if func handles arrays as arguments. The default value is True.


```python
from scipy import integrate
import numpy as np

f= lambda x: x**2

int_result= integrate.quadrature(f, 0.0, 1)

print(int_result)
```

## **Legendre Polynomails** 


```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.special import legendre
from scipy import integrate 

def legpoly(x,n):
    leg=legendre(n)
    P_n = leg(x) 
    return P_n

def di_legendre(n,m):
    leg=legendre(n)*legendre(m)
    P_nm=leg 
    return P_nm

a=int(input("Enter value of n = "))
b=int(input("Enter value of m = ")) 

#print(a,b,di_legendre(x,a,b))
#print(legendre(3))
result=integrate.quad(di_legendre(a,b),-1,1)

print(a,b,result) 

x=np.linspace(-1,1,100)

n=5

for i in range(n):
    plt.plot(x,legpoly(x,i),label='$P_{}(x)$'.format(i))

plt.title("Legendre Polynomails")
plt.xlabel("x")
plt.ylabel("$P_n(x)$")
plt.legend(loc='lower right')
plt.show()
```

## Triangle wave

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt 

from scipy.signal import square 
from scipy.integrate import quad 
from math import * 

x = np.arange(0, 101)

def tri(x):
    res = x % 50
    if res < 25:
        return res
    else:
        return 50 - res

plt.plot(x, list(map(tri, x)))
```

## Sawtooth wave

```python
import numpy as np
from scipy.signal import sawtooth
import matplotlib.pyplot as plt 

from scipy.integrate import quad 

x = np.linspace(0, 2*np.pi, 500)

def sawt(x):
    return sawtooth(x,width=0.5)

y = sawt(x) 

fc=lambda x:sawt(x)*cos(i*x) 
fs=lambda x:sawt(x)*sin(i*x)

An=[] 
Bn=[] 

n=8
i=0
a0=quad(fc,-np.pi,np.pi)[0]*(1/np.pi)
sum=a0/2

for i in range(n):
    an=quad(fc,-np.pi,np.pi)[0]*(1/np.pi) 
    An.append(an) 
    bn=quad(fs,-np.pi,np.pi)[0]*(1/np.pi) 
    Bn.append(bn) 

for i in range(n):
    sum= sum + An[i]*np.cos(i*x) + Bn[i]*np.sin(i*x) 

plt.plot(x,sum,'r--') 

plt.plot(x,y,'b')
plt.show() 
```

## Half wave rectifier

```python
import numpy as np
import matplotlib.pyplot as plt 

from math import * 


def ff(t,period):
    tmp=[] 
    for i in t:
        if (i >= period/2): 
            result = 0
        else:
            result = np.sin(i) 
        tmp.append(result) 
#    print(t,tmp) 
    return tmp  

x=np.linspace(0,2*np.pi,100) 
period = 2*np.pi 

y=ff(x,period) 
plt.plot(x,y, color='red')
plt.show() 
```
