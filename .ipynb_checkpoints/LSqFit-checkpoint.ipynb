# **Least Square Fit** 

```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy.linalg import solve

x=np.array([1,2,3,4,5])
y=np.array([3.1,5.9,7.2,10.0,11.3]) 

# Computing the sums

n   = x.size 
sx  = np.sum(x)
sy  = np.sum(y)
sx2 = np.sum(x**2)
sxy = np.sum(x*y)

# construct matrices 
A = np.array([[n,sx],[sx,sx2]]) 
B = np.array([sy,sxy]) 

# Solutions of equations using linear algebra module 

C = solve(A,B) 
a0,a1 = C

# Plot 
yfit = a0 + a1*x
plt.plot (x,y, 'o') 
plt.plot(x,yfit) 
plt.show() 
```
